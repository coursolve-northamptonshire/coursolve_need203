```python
import collections
import stopwords
import re

stop_dict = {w: True for w in stopwords.WORDS}
re_handtagurl = re.compile(r'^.*(\@|\#|https?\:\/\/).*$', re.I)
re_nonascii = re.compile(r'[^\x00-\x7F]',re.I)
re_punc_split = re.compile(r'[\.\,\:\\\/]')
re_strip_chars = re.compile(r'[^0-9A-Za-z\-\_]')

min_len = 3

def get_documents(tweets_df, min_len):
    ''' Takes a dataframe of tweets & returns an array of word lists i.e. documents
    Words added to the list are filtered by this criteria:
    1. Should not be hashtags/handles/URLs
    2. Should be Ascii printable characters
    3. words concatenated with simple punctuations are split (except hyphenated or underscored).
    4. Are not stop words.
    5. Length of the word >= min_len.
    '''
    def get_words(text):
        ''' Takes a tweet and tokenizes it, returning all words that occur atleast more than once.
        '''
        words_in_doc = []
        def junk_filter(words):
            def sub_word_tokenizer(w):
                if not re_handtagurl.match(w) and not re_nonascii.match(w):
                    for w in map(lambda w : re_strip_chars.sub('', w),
                                 re_punc_split.split(w)):
                        if w not in stop_dict and len(w) >= min_len:
                            yield w
            for w in words:
                for sw in sub_word_tokenizer(w.strip()):
                    yield sw
                
        words_in_doc = [ w for w in junk_filter(text.lower().split()) ]        
        return words_in_doc
    
    documents = tweets_df["status"].apply(get_words).tolist()
    
    return documents

documents = get_documents(dataframe, min_len)
documents

    
```

```python

from gensim import corpora, models, similarities

id2word = corpora.Dictionary(documents)

mm = [id2word.doc2bow(document) for document in documents]

mm
```

```python
lda = models.ldamodel.LdaModel(corpus=mm, id2word=id2word, num_topics=50, update_every=1, chunksize=10000, passes=5)
```
