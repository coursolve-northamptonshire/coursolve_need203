```python
import collections
import stopwords
import re

stop_dict = {w: True for w in stopwords.WORDS}
re_handtagurl = re.compile(r'^.*(\@|\#|https?\:\/\/).*$', re.I)
re_nonascii = re.compile(r'[^\x00-\x7F]',re.I)

min_len = 3

def getVocabulary(tweets_df, min_len):
    corpus_counter = collections.Counter()
    def count_words(text):
        doc_counter = collections.Counter()
        for w in text.split():
            w = w.strip().lower()
            if w not in stop_dict and len(w) >= min_len:
                if not re_handtagurl.match(w):
                    if not re_nonascii.match(w):
                        doc_counter[w] += 1
                        corpus_counter[w] += 1
                    else:
                        continue
                else:
                    continue
            else:
                continue
        return doc_counter
    #doc_counters = []
    doc_counters = tweets_df["status"].apply(count_words)
    return (doc_counters, corpus_counter)


(doc_counters, corpus_counter) = getVocabulary(dataframe, min_len)
doc_counters

    
```
